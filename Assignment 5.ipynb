Question 1: Age Group Classification 

Write a Python program that takes the age of a person as input and classifies them into one of the following age groups:

"Child" if the age is between 0 and 12 (inclusive).

"Teenager" if the age is between 13 and 19 (inclusive).

"Adult" if the age is between 20 and 64 (inclusive).

"Senior" if the age is 65 or older.


```python
# Age Group Classification

# Take age as input
age = int(input("Enter your age: "))

# Classify the age group
if 0 <= age <= 12:
    print("Child")
elif 13 <= age <= 19:
    print("Teenager")
elif 20 <= age <= 64:
    print("Adult")
elif age >= 65:
    print("Senior")
else:
    print("Invalid age")
```

Q2. Write a Python program that takes three numbers as input and determines the largest among them. Also, check if any or all of the numbers are equal."

```python
# Take input
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

# Check equality
if num1 == num2 == num3:
    print("All numbers are equal")
elif num1 == num2 or num2 == num3 or num1 == num3:
    print("Some numbers are equal")

# Find largest number
if num1 >= num2 and num1 >= num3:
    print("Largest number is:", num1)
elif num2 >= num1 and num2 >= num3:
    print("Largest number is:", num2)
else:
    print("Largest number is:", num3)
```

Q4. Write a Python program to store and display the details of a book.

The program should ask the user to input the following information:

Title of the book

Author’s name

Year of publication

Price of the booK

Store the entered details in a tuple.

Finally, display the book details in a proper format as shown below:

Expected Output:

Title: Harry Potter

Author: JK Rowlings

Year of Publication: 2000

Price: $180.0

```python
# Take input
title = input("Enter book title: ")
author = input("Enter author name: ")
year = int(input("Enter year of publication: "))
price = float(input("Enter book price: "))

# Store in tuple
book = (title, author, year, price)

# Display details
print("\nTitle:", book[0])
print("Author:", book[1])
print("Year of Publication:", book[2])
print("Price: $", book[3])
```

Q5. Write a program to check if a user’s chosen subject is available in the list of offered subjects.

Instructions:

Given a list of subjects, ask the user to input a subject.

Display whether the subject is available or not.

Expected Output:

Enter the subject you are interested in:  art

art is available.

```python
# List of subjects
subjects = ["Math", "English", "Science", "Art", "Computer"]

# Take input
subject = input("Enter the subject you are interested in: ")

# Check availability
if subject.lower() in [s.lower() for s in subjects]:
    print(subject, "is available.")
else:
    print(subject, "is not available.")
```

Q9. Write a Python program to check whether the entered letter is a vowel, a consonant, or invalid input.

The program should take a single character as input. If the character is a vowel (a, e, i, o, u in both upper and lower case), it will print "Vowel". If it is an alphabet but not a vowel, it will print "Consonant". Otherwise, it will print "Please enter valid alphabet".

Expected Output:

Enter a single letter:  a

a is a Vowel

---

Enter a single letter:  E

E is a Vowel

---

Enter a single letter:  z

z is a Consonant

---

Enter a single letter:  7

Please enter valid alphabet


```python
# Take input
letter = input("Enter a single letter: ")

# Check input
if len(letter) == 1 and letter.isalpha():
    if letter.lower() in "aeiou":
        print(letter, "is a Vowel")
    else:
        print(letter, "is a Consonant")
else:
    print("Please enter valid alphabet")
```
